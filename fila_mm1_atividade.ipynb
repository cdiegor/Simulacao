```python
import numpy as np
import simpy as sp
```

```python
# Variáveis de sistema
n = 0
taxa_chegada = 2
taxa_atendimento = 3
horizonte = 300
```

```python
from numpy.lib.nanfunctions import nanargmax
def chegada(env, atendentes, lam, mu):
  id_cliente = 0
  while True:
    yield env.timeout(np.random.exponential(1/lam))
    id_cliente = id_cliente + 1
    print("Alguem chegou \t (n = ", n, " em t = ", str(env.now), ")")
    env.process(atendimento(env, atendentes, mu, id_cliente, n))
```

```python
def atendimento(env, atendentes, mu, id_cliente, n):

    at = atendentes.request()
    yield at
    yield env.timeout(np.random.exponential(1/mu))
    print("Alguem saiu   \t (n = ", n, " em t = ", str(env.now), ")")
    atendentes.release(at)
```

```python
env = sp.Environment()
atendentes = sp.Resource(env, capacity=2)
env.process(chegada(env, atendentes, taxa_chegada, taxa_atendimento))
env.run(until=horizonte)
print("Fim da simulação")
```

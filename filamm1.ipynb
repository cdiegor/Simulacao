```python
import numpy as np
import simpy as sp
```

```python
# Variáveis de sistema
n = 0
taxa_chegada = 2
taxa_atendimento = 3
horizonte = 300
```

```python
def chegada(env):
  global n
  while True:
    yield env.timeout(np.random.exponential(1/taxa_chegada))
    n = n + 1
    print("Alguem chegou \t (n = ", n, " em t = ", str(env.now), ")")
    if (n == 1):
      env.process(atendimento(env))

```

```python
def atendimento(env):
  global n
  while (n > 0):
    yield env.timeout(np.random.exponential(1/taxa_atendimento))
    n = n - 1
    print("Alguem saiu   \t (n = ", n, " em t = ", str(env.now), ")")
```

```python
env = sp.Environment()
env.process(chegada(env))
if (n > 0):
  env.process(atendimento(env))
env.run(until=horizonte)
print("Fim da simulação")
```

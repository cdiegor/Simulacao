```python
import numpy as np
import simpy as sp
```

```python
def chegada(env):
  global n
  global custo
  
  while True:
    yield env.timeout(np.random.exponential(media_chegada))

    q = np.random.binomial(5, 1/2)
    r = min(n, q)
    if (n > 0):
      n = n - r
      #custo = custo - r*valor_unitario
      if (n + r >= s and n < s):
        env.process(recarga(env, S-n))
    custo = custo + (q-r)*valor_unitario

    print("Cliente chegou \t (q =", q, "r = ", r, "n = ", n, "custo = ", custo, ")")
```

```python
def recarga(env, R):
  global n
  global custo

  print("Recarga agendada \t (R = ", R, ")")

  yield env.timeout(np.random.exponential(media_recarga))

  n = n + R
  custo = custo + recarga_fixo + recarga_unitario*R

  print("Recarga efetuada \t (n = ", n, "custo = ", custo, ")")
```

```python
# VariÃ¡veis do sistema
S = 500
s = 150
n = S
custo = 0

media_chegada = 1/20
media_recarga = 5

media_compra = 3
variancia_compra = 1
valor_unitario = 10

recarga_fixo = 100
recarga_unitario = 5
```

```python
env = sp.Environment()
env.process(chegada(env))
env.run(until=30)
print("Fim da simulaÃ§Ã£o")
```
